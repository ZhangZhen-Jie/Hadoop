# Step.1 master、slaver1、slaver2都要執行以下指令
## 下載 jdk

```console
[root@localhost ~]# yum update
[root@localhost ~]# yum install wget git ntp
[root@localhost ~]# cd /tmp; wget https://mail-tp.fareoffice.com/java/jdk-8u144-linux-x64.rpm
[root@localhost ~]# cd ~
```
## 安裝 jdk (建議使用 Oracle java)

```console
[root@localhost ~]# rpm -ivh /tmp/jdk-8u144-linux-x64.rpm
```

## 變更內建java為(Oracle java)

- (※新安裝的系統應該只會有內建java與jdk-8u144-linux-x64，兩種)

```console
[root@localhost ~]# update-alternatives --config java
有 2 個程式提供「java」。

  選擇        指令
-----------------------------------------------
*+ 1           java-1.8.0-openjdk.x86_64 (/usr/lib/jvm/java-1.8.0-openjdk-1.8.0.222.b10-0.el7_6.x86_64/jre/bin/java)
   2           /usr/java/jdk1.8.0_144/jre/bin/java

請輸入以保留目前的選擇[+]，或輸入選擇號碼：
```
- 看到```請輸入以保留目前的選擇[+]，或輸入選擇號碼：```

# Step.2 master、slaver1、slaver2都要執行以下指令

## 建立軟連結
```console
[root@localhost ~]# ln -s /usr/java/jdk1.8.0_144/ /usr/java/java
```
## 設定環境目錄
- 使用vi開啟profile(系統環境目錄)
- ```wq```: 存檔離開
- ```q```: 未更動內容，直接離開
- ```q!```: 不存檔，強制離開

```console
[root@localhost ~]# vi /etc/profile
```
- 使用vi進入後，輸入
    1. ```Shift + g```: 游標到最後一行
    2. ```o```: 於下一行進入```insert mode```

---
- 於檔案中的最末段加入以下內容
```console
export JAVA_HOME=/usr/java/java
export JRE_HOME=$JAVA_HOME/jre
export CLASSPATH=.:$JAVA_HOME/lib/dt.jar:$JAVA_HOME/lib/tools.jar:$JRE_HOME/lib/rt.jar
export PATH=$PATH:$JAVA_HOME/bin 
```


# Step.3 master、slaver1、slaver2都要執行以下指令
## 設定安全性
```console
[root@localhost ~]# setenforce 0 
```

## 永久停止selinux安全性

```console
[root@localhost ~]# sed -i '/SELINUX/s/enforcing/disabled/' /etc/selinux/config 
```
- ```sed```指令解說 : https://thecodeninja.net/2019/02/keyword-based-inline-search-and-replace-with-sed/
    1. ```SELINUX``` : 游標移至，keyword字串「SELINUX」之後
    2. ```s``` : 搜尋「search」
    3. ```enforcing``` : search_string
    4. ```disabled``` : replace_string
    5. ```/etc/selinux/config``` : 檔案位置

## 停止防火牆
```console
[root@localhost ~]# systemctl disable firewalld
[root@localhost ~]# systemctl stop firewalld
```

## 設定SSH登入免詢問
```console
[root@localhost ~]# echo 'StrictHostKeyChecking no' >> /etc/ssh/ssh_config
[root@localhost ~]# systemctl restart sshd
```

## 產生 keygen
```console
[root@master ~]# ssh-keygen -t rsa -f ~/.ssh/id_rsa -P ""
[root@master ~]# cat ~/.ssh/id_rsa.pub >> ~/.ssh/authorized_keys
[root@master ~]# ssh localhost
```
- ```ssh localhost```: 確認登入自己，還需不需要確認密碼(設定完之後，應該是不需要密碼，才可以繼續)
- ```ssh master```: 確認登入master(本機別名)，還需不需要確認密碼(設定完之後，應該是不需要密碼，才可以繼續)
- ```exit```或```Ctrl + D```: 登出 or 離開bash
- ```users```: 確認本機有多少個使用者(用來確認自已重複登入自己本機的狀態)

## 複製 key 到其他兩台
```console
[root@master ~]# ssh-copy-id -i ~/.ssh/id_rsa.pub root@slaver1
[root@master ~]# ssh-copy-id -i ~/.ssh/id_rsa.pub root@slaver2
[root@master ~]# ssh slaver1
[root@slaver1 ~]# exit
[root@master ~]# ssh slaver2
[root@slaver2 ~]# exit
```
- ```ssh slaver1```: 確認登入slaver1，還需不需要確認密碼(設定完之後，應該是不需要密碼，才可以繼續)
- ```ssh slaver2```: 確認登入slaver1，還需不需要確認密碼(設定完之後，應該是不需要密碼，才可以繼續)

# Step12. 產生 Hadoop 連結檔(在master、slaver1、slaver2做)

# 關掉/ 切換 NN active 服務(模擬Active主機失效，與回復負載平衡)

## 在 Active 主機上，關閉Active服務(Master順序上應該是active狀態)

```console
[root@master ~]# hadoop-daemon.sh stop namenode
[root@master ~]# hdfs haadmin -getServiceState nn2
```

## 狀況:Master已經停止服務，所以去slaver1建立文檔
- 查看NN(Name Node)服務狀態，已經不能進入
> http://192.168.225.130:50070

```console
[root@slaver1 ~]# hadoop fs -mkdir /test
[root@slaver1 ~]# hadoop fs -ls /
```
- 輸入以上指令後，服務還可以運作，只是查看slaver1(nn2)的NN服務狀態
> http://192.168.225.131:50070
- 已經從 ```'slaver1:8020'(standby)```轉換成 ```'slaver1:8020'(active)```

## 狀況:Master已經修復，啟動Master

```console
[root@master ~]# hadoop-daemon.sh start namenode
```
- master的NN服務狀態 ```'master:8020'(standby)```


## 切換 NN 狀態(模擬回復負載平衡)

```console
[root@master ~]# hdfs haadmin -failover nn2 nn1 
```
- 將master(nn1)的NN服務狀態設定為 ```'master:8020'(active)```

# 關閉/ 開啟方式

## 關閉服務
- ```{PID}```為服務(zk)的process ID

```console
[root@master ~]# stop-all.sh
[root@master ~]# kill -9 {PID} 
```

---

## 開啟服務

### 啟動 ZK (每一台)

```console
[root@master ~]# /opt/zookeeper/bin/zkServer.sh start
```
### 啟動 Journal(每一台)

```console
[root@master ~]# hadoop-daemon.sh start journalnode
```

### 啟動 hdfs，會自動帶起 NameNode
```console
[root@master ~]# start-all.sh
```
